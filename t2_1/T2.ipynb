```python
# products
df_products_raw = pd.read_parquet(path+filenames[1])
df_products_raw = df_products_raw[df_products_raw['product_locale']=='us']
df_products_raw.head()
```

```python
cols_to_keep_from_products = ['product_id','product_title','product_description','product_bullet_point','product_brand','product_color']
cols_to_keep_from_examples = ['query','product_id']
df_products = pd.merge(df_products_raw[cols_to_keep_from_products].drop_duplicates(), df_examples_raw[cols_to_keep_from_examples].drop_duplicates() ,on = ['product_id'])

df_products = df_products.sample(n=50,random_state=42)
print(df_products.shape)
df_products.head()
```

```python
from transformers import AutoTokenizer, TFAutoModel

model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = TFAutoModel.from_pretrained(model_ckpt, from_pt=True)
```

```python
def concatenate_text(row):
    return {
        "text": str(row['product_title']) + "\n" +
        str(row['product_description']) + "\n" +
        str(row['product_bullet_point']) + "\n" +
        str(row['product_brand']) + "\n" +
        str(row['product_color']) + "\n" +
        str(row['query']) + "\n"
    }
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="tf"
    )
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
from datasets import Dataset

products_dataset = Dataset.from_pandas(df_products)
products_dataset = products_dataset.map(concatenate_text)
```

```python
embedding = get_embeddings(products_dataset["text"][0])
embedding.shape
```

```python
embeddings_dataset = products_dataset.map(
    lambda x: {"embeddings": get_embeddings(x["text"]).numpy()[0]}
)
```

```python

import faiss
embeddings_dataset.add_faiss_index(column="embeddings")
```

```python
question = "Men's Striped Shirt"
question_embedding = get_embeddings([question]).numpy()
question_embedding.shape
```

```python
scores, samples = embeddings_dataset.get_nearest_examples(
    "embeddings", question_embedding, k=5
)
```

```python
import pandas as pd

samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values("scores", ascending=False, inplace=True)
```

```python
for _, row in samples_df.iterrows():
    print(f"COMMENT: {row.comments}")
    print(f"SCORE: {row.scores}")
    print(f"TITLE: {row.title}")
    print(f"URL: {row.html_url}")
    print("=" * 50)
    print()
```
